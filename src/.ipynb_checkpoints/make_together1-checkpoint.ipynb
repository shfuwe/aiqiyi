```python
def ana(path):
    f=open(path,encoding='utf-8', mode='r')
    line = f.readlines()

    import ast
    li=[]
    for i in line:
        i=i.replace('\n','')
        i = i.split('\t')
        if i[1]!= 'emotion':
            i[1]=ast.literal_eval(i[1])
        li.append(i)
    return li
#         print(i)
    
```

```python
re_l=['result_1636180246.4086099_0_0.7081_0.7007','result_1636556972.861262_2_0.6886_0.6942','result_1636619329.723694_1_0.6912_0.7004',
     'result_1636974376.1919332_1_0.7001_0.6911','result_1637069358.4556513_3_0.6593_0.64','result_1637087331.853575_3_0.6959_0.6943'
     ]
a_l=[ana('../data/'+i+'.txt') for i in re_l]
```

```python
import numpy as np

def toupiao(a):
    piao=np.zeros(4)
    for i in a:
        piao[i]+=1
    max_p=max(piao)
    num_=0
    sum_=0
    for i in range(len(a)):
        if piao[a[i]]==max_p:
            sum_+=a[i]
            num_+=1
#     print(sum_,num_)
    return round(sum_/num_)
    
```

```python
import time 
end=time.time()
ff = open('../data/result_6_'+str(end)+'.txt', encoding='utf-8', mode='w')
ff.write('id\temotion\n')

for i in range(1,len(a_l[0])):
    i_l=[]
    for j in range(6):
#         print(i,j)
        a=[k[i][1][j] for k in a_l]
        i_l.append(toupiao(a))
    
    s=''
    for j in i_l:
        s+=str(j)
        s+=','
    s=s[:-1]
    
    ff.write(a_l[0][i][0]+'\t'+s)
    ff.write("\n")
    
ff.close()
```
