```python
def ana(path):
    f=open(path,encoding='utf-8', mode='r')
    line = f.readlines()

    import ast
    li=[]
    for i in line:
        i=i.replace('\n','')
        i = i.split('\t')
        if i[1]!= 'emotion':
            i[1]=ast.literal_eval(i[1])
        li.append(i)
    return li
#         print(i)
    
```

```python
a_6823=ana('../data/ttt/0.6823_result_1636180246.4086099_0_0.7081_0.7007.txt')
a_6761=ana('../data/ttt/0.6761_result_1636362793.6436496_0_0.8608_0.8349.txt')
a_6781=ana('../data/ttt/0.6781_result_1636556972.861262_2_0.6886_0.6942.txt')


```

```python
def _3to1(a,b,c):
    if a==b and b==c:
        return a
    if a==b and b!=c:
        return a
    if a==c and a!=b:
        return a
    if b==c and a!=b:
        return b
    if a!=b and a!=c and b!=c:
        return round((a+b+c)/3)
```

```python
import time 
end=time.time()
ff = open('../data/result_3_'+str(end)+'.txt', encoding='utf-8', mode='w')
ff.write('id\temotion\n')

for i in range(1,len(a_6823)):
    i_l=[]
    for j in range(6):
        i_l.append(_3to1(a_6823[i][1][j],a_6761[i][1][j],a_6781[i][1][j]))
    
    s=''
    for j in i_l:
        s+=str(j)
        s+=','
    s=s[:-1]
    
    ff.write(a_6823[i][0]+'\t'+s)
    ff.write("\n")
```
