```python
def ana(path):
    f=open(path,encoding='utf-8', mode='r')
    line = f.readlines()

    import ast
    li=[]
    for i in line:
        i=i.replace('\n','')
        i = i.split('\t')
        if i[1]!= 'emotion':
            i[1]=ast.literal_eval(i[1])
        li.append(i)
    return li
#         print(i)
    
```

```python
re_l=['result_1636619329.723694_1_0.6912_0.7004','result_1636974376.1919332_1_0.7001_0.6911','result_1637309018.2075813_2_0.6927_0.7114'
     ]
a_l=[ana('../data/'+i+'.txt') for i in re_l]
```

```python
import numpy as np

def toupiao(a):
#     print(len(a))
    sum_=0
    for i in a:
        sum_+=i
#     if sum_/len(a)>2:
#         print(sum_/len(a))
    return sum_/len(a)
    
```
